```python
from scipy import special

# Calcular el factorial de un número usando SciPy
n = 5
factorial_sp = special.factorial(n, exact=True)
print(f"El factorial de {n} es {factorial_sp}")
```

```python
def nap(n):
  return np.sum(1/special.factorial(i, exact=True) for i in range(n+1))

nap(9)
```

```python
import numpy as np

# Calcular el factorial de un número usando NumPy
n = 5
factorial_np = np.math.factorial(n)

print(f"Factorial de {n} usando NumPy: {factorial_np}")
```

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)

factorial(5)
```
